```python
#Univariate Analysis
def discrete_univariate(df,feature):
    total=len(df[feature])
    count=df[feature].nunique()
    plt.figure(figsize=(count+1,5))
    ax=sns.countplot(data=df,x=feature,order=df[feature].value_counts().index.sort_values(ascending=False))
    for a in ax.patches:
        label="{:.1f}%".format(100*a.get_height()/total)
        x=a.get_x()+a.get_width()/2
        y=a.get_height()
        ax.annotate(label,(x,y),ha='center',va='center',xytext=(0,5),size=8,textcoords="offset points")
    plt.show()
```

```python
discrete_univariate(df,'CLASS_LABEL')
```

```python
#Bivariate Analysis
def bivariate_analysis(df,feature_list):
    plt.figure(figsize=(15,10))
    for i,var in enumerate(feature_list):
        plt.subplot(3,2, i+1)
        sns.boxplot(df['CLASS_LABEL'],df[var], palette='viridis')
        plt.tight_layout()
        plt.title('Class Label vs'+var)
    plt.show()
```

```python
bivariate_analysis(df, ['NumDots','PathLevel','UrlLength','SubdomainLevel','NumNumericChars','PathLength'])
```

```python
def heatmap(df,start,end):
    new_df=pd.DataFrame(df.iloc[:,start:end])
    plt.figure(figsize=(10,8))
    sns.heatmap(new_df.corr(),annot=True,fmt='.3f')
    
```

# Splitting Dataset into training and testing dataset

```python
x=df.drop(columns=['id','CLASS_LABEL'],axis=1)
y=df['CLASS_LABEL']
```

```python
#training set=80% and testing set=20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

# XGBoost

```python
#Model fitting
from xgboost import XGBClassifier
clf1=XGBClassifier()
clf1.fit(x_train,y_train)
```

```python
#Prediction for testing values
y_pred1=clf1.predict(x_test)
y_pred1
```

```python
#Prediction for training values
y_pred2=clf1.predict(x_train)
y_pred2
```

```python
#Performance Evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,recall_score,precision_score,accuracy_score
```

```python
#MSE
mean_squared_error(y_test,y_pred1)
```

```python
#MAE
mean_absolute_error(y_test,y_pred1)
```

```python
#RMSE
mean_squared_error(y_test,y_pred1,squared=False)
```

```python
#Recall
recall_score(y_test,y_pred1)
```

```python
#Precision
precision_score(y_test,y_pred1)
```

```python
#Accuracy
print("Accuracy score : ",accuracy_score(y_test,y_pred1))
```

```python
#Performance for training dataset
accuracy_score(y_train,y_pred2)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
print("Confusion matrix for testing dataset : \n",confusion_matrix(y_test,y_pred1))
```

```python
print("Confusion matrix for training dataset : \n",confusion_matrix(y_train,y_pred2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
